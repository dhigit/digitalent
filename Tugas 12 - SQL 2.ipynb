```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def runsql(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# One to many relationshipÂ¶ 

```python
#    Report the account representative for each customer.
runsql("""
    SELECT 
        a.customerName, b.*
    FROM
        customers a
        INNER JOIN employees b
            ON a.salesRepEmployeeNumber = b.employeeNumber
""")
    
```

```python
#    Report total payments for Atelier graphique.
runsql ("""
    SELECT 
        SUM(amount)
    FROM
        customers 
        INNER JOIN payments 
            USING (customerNumber)
    WHERE
        customerName = 'Atelier graphique'
""")
```

```python
#    Report the total payments by date
runsql ("""
    SELECT 
        paymentDate, SUM(amount)
    FROM
        payments
    GROUP BY
        paymentDate
""")
```

```python
#    Report the products that have not been sold.
runsql ("""
    SELECT 
        a.*
    FROM
        products a
        LEFT JOIN orderdetails b
            USING (productCode)
    WHERE
        orderNumber IS NULL
""")
```

```python
#    List the amount paid by each customer.
runsql ("""
    SELECT 
        customerName, SUM(amount)
    FROM
        customers
        LEFT JOIN payments
            USING (customerNumber)
    GROUP BY
        customerNumber
""")
```

```python
#    How many orders have been placed by Herkku Gifts?
runsql ("""
    SELECT 
        customerName, COUNT(*)
    FROM
        customers
        INNER JOIN orders
            USING (customerNumber)
    WHERE
        customerName = 'Herkku Gifts'
    GROUP BY
        customerNumber
""")
```

```python
#    Who are the employees in Boston?
runsql ("""
    SELECT 
        a.*
    FROM
        employees a
        INNER JOIN offices
            USING (officeCode)

    WHERE
        city = 'Boston'
""")
```

```python
#    Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.
runsql ("""
    SELECT 
        customerName, SUM(amount) AS totalPayment
    FROM 
        customers
        INNER JOIN payments
            USING (customerNumber)
    GROUP BY
        customerNumber
    ORDER BY
        totalPayment DESC
""")  
```

```python
#List the value of 'On Hold' orders.
runsql ("""
    SELECT 
        orderNumber, STATUS, (quantityOrdered * priceEach) AS VALUE
    FROM 
        orders
        INNER JOIN orderdetails
            USING (orderNumber)
    WHERE
        STATUS = 'On Hold'
    GROUP BY
        orderNumber
""")
```

```python
#Report the number of orders 'On Hold' for each customer.
runsql ("""
    SELECT 
        customerNumber, customerName, status, COUNT(*)
    FROM 
        orders
        INNER JOIN customers
            USING (customerNUmber)
    WHERE
        STATUS = 'On Hold'
    GROUP BY
        customerNumber
""")
```

# Many to many relationshipÂ¶ 

```python
#   List products sold by order date.
runsql ("""
    SELECT 
        orders.orderDate, products.*
    FROM 
        orders
        INNER JOIN orderdetails
            USING (orderNumber)
        INNER JOIN products
            USING (productCode)
""")
```

```python
#   List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
runsql ("""
    SELECT 
        orders.orderDate
    FROM 
        orders
        INNER JOIN orderdetails
            USING (orderNumber)
        INNER JOIN products
            USING (productCode)
    WHERE
        productName = '1940 Ford Pickup Truck'
    GROUP BY
        orderDate
    ORDER BY
        orders.orderDate DESC
""")
```

```python
#    List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?
runsql ("""
    SELECT 
        `customerName`, `orderNumber`, (quantityOrdered * priceEach) AS orderValue
    FROM 
        customers
        INNER JOIN orders
            USING (customerNumber)
        INNER JOIN orderdetails
            USING (ordernumber)
    HAVING
        orderValue > 25000
""")
```

```python
#    Are there any products that appear on all orders?
runsql ("""
    SELECT 
        productName, COUNT(*) AS productOrderedQty
    FROM 
        orderdetails
        INNER JOIN products
            USING (productCode)
    GROUP BY 
        productCode
    HAVING
        productOrderedQty = (SELECT COUNT(*) FROM orders)
""")
```

```python
#    List the names of products sold at less than 80% of the MSRP.

runsql("""
    SELECT 
        productName
    FROM 
        products
        INNER JOIN orderdetails
            USING (productCode)
    WHERE
        priceEach < MSRP*0.8
    GROUP BY
        productName
""")
```

```python
#    Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)
runsql ("""
    SELECT 
        productName, buyPrice, priceEach
    FROM 
        products
        INNER JOIN orderdetails
            USING (productCode)
    WHERE
        priceEach >= buyPrice*2
""")
```

```python
#    List the products ordered on a Monday.
runsql ("""
    SELECT 
        products.*
    FROM 
        orders
        INNER JOIN orderdetails
            USING (orderNumber)
        INNER JOIN products
            USING (productCode)
    WHERE
        DAYNAME(orderDate) = 'Monday'
    GROUP BY
        productCode
""")
```

```python
#    What is the quantity on hand for products listed on 'On Hold' orders?

runsql ("""
    SELECT 
        productName, SUM(quantityOrdered)
    FROM 
        orders
        INNER JOIN orderdetails
            USING (orderNumber)
        INNER JOIN products
            USING (productCode)
    WHERE
        STATUS = 'On Hold'
    GROUP BY 
        productCode
""")
```
