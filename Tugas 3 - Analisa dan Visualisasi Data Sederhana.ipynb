# Deskripsi

**Indeks Pembangunan Manusia (IPM)** atau Human Development Index (HDI) adalah sebuah indeks yang menyatakan standar hidup (kualitas hidup) di suatu negara. Semakin besar IPM, maka semakin baik standar hidup pada negara tersebut.

**Rasio Gini** dalam ekonomi menunjukkan tingkat kemerataan pendapatan masyarakat dalam suatu negara. Dimana, rasio gini 0 menunjukkan pendapatan yang paling merata (most equal), dan 1 pendapatan yang paling tidak merata (most inequal). 

Data diambil dari https://data.go.id. Data IPM berisi kode_provinsi, nama_provinsi, tahun, dan indeks_pembangunan_manusia. Sedangkan data Rasio Gini berisi kode_provinsi, nama_provinsi, tahun, dan rasio_gini. Data IPM dan Rasio Gini memiliki 2 key yang sama, yaitu kode_provinsi dan tahun. Kedua key tersebut dapat dimanfaatkan untuk menggabungkan kedua data dengan akurat.

Dengan memanfaatkan kedua data tersebut, dapat dilihat bagaimana korelasi antara IPM dan Rasio Gini.


## Load Dataset

```python
import pandas as pd

data_ipm = pd.read_csv('tugas/index_pembangunan_manusia.csv', delimiter = ',', encoding = "utf-8")
data_gini = pd.read_csv('tugas/gini_ratio.csv', delimiter = ',', encoding = "utf-8")
```

```python
data_ipm.dropna().sample(5)
```

```python
data_gini.dropna().sample(5)
```

## Statistik Data Index Pembangunan Manusia

```python
data_ipm.groupby('tahun').describe()['indeks_pembangunan_manusia']
```

## Statistik Data Index Gini

```python
data_gini.groupby('tahun').describe()['rasio_gini']
```

## Menggabungkan Data Menjadi 1 Dataframe

```python
# gabungkan data menjadi 1 dataframe berdasarkan kode_provinsi dan tahun
# lalu hapus row yang tidak ada IPM atau Rasio Gini

combined_data = data_ipm.merge(data_gini, how='inner', on=['kode_provinsi', 'tahun']).dropna()
combined_data.sample(5)
```

## Statistik data yang telah digabungkan

```python
combined_data.groupby('tahun').describe()[['indeks_pembangunan_manusia', 'rasio_gini']]
```

## Korelasi Antara Indeks Pembangunan Manusia dan Rasio Gini

```python
# kalkulasi mean berdasarkan tahun
vis = combined_data.groupby('tahun')[['indeks_pembangunan_manusia','rasio_gini']].mean()
vis.corr()
```

## Visualisasi Data

```python
import seaborn as sea

sea.lmplot(x='indeks_pembangunan_manusia', y='rasio_gini', data=vis)
```

## Kesimpulan
Berdasarkan perhitungan dan visualisasi data di atas, dapat dilihat bahawa IPM berkorelasi positif dengan Rasio Gini (r=0,948). Hal ini menujukkan bahwa jika terjadi kenaikan pada IPM, maka Rasio Gini juga ikut naik. Analisa lebih jauh dapat dikatakan bahwa dari tahun ke tahun kualitas hidup masyarakat Indonesia meningkat, namun kesenjangan penghasilan juga semakin renggang. Data yang digunakan pada analisis ini adalah data dari tahun 1996 hingga 2010. namun, dilihat dari grafik regresi linier yang dipaparkan di atas, trend ini diprediksi akan tetap sama hingga saat ini. Kesimpulan lebih jauh dapat dicapai jika terdapat data hingga tahun 2018.
